```python
filtro5 = df.tail(15)
filtro5
# tail es para los últimos registros
```

```python
filtro6 = df[df['COSTO_TOTAL_CALCULADO'] > 1000]
filtro6
# Suma de registros mayores a 1000
```

```python
filtro6.to_csv('filtro6.csv')
files.download('filtro6.csv')
```

```python
filtro7 = df[(df['NOMBRE_VENDEDOR'] == 'LETICIA RAMIREZ HERNANDEZ') & (df['CANT'] > 10)]
filtro7
# Filtro de las ventas de más de 10 artículos de Leticia Ramírez
```

```python
filtro7.to_csv('filtro7.csv')
files.download('filtro7.csv')
```

```python
filtro8 = df[(df['NOMBRE_VENDEDOR'] == 'ALFREDO CANELA') | (df['CANT'] > 50)]
filtro8
# Con este filtro se ven todas las ventas de más de 50 artículos o de Alfredo Canela
```

```python
filtro8.to_csv('filtro8.csv')
files.download('filtro8.csv')
```

```python
filtro9 = df[~(df['CANT'] < 50) & ~(df['PRECIO_UNITARIO'] < 1000)]
filtro9
# Ventas donde la cantidad no sea menor que 50 y donde el precio unitario no sea menor que $1000
```

```python
filtro9.to_csv('filtro9.csv')
files.download('filtro9.csv')
```
